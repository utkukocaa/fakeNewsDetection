```python
from textblob import TextBlob
from sklearn import model_selection,preprocessing, linear_model, naive_bayes
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn import decomposition, ensemble
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

import pandas, numpy, textblob,string
```

```python
import pandas as pd
data_fake = pd.read_csv("fake.csv")
data_fake.head()
```

```python
data_fake["status"] = "fake"
data_fake.dropna()
```

```python
df_fake = pd.DataFrame()
df_fake["title"]= data_fake["title"].apply(str)
df_fake["status"] = data_fake["status"]
df_fake.head()
```

```python
#buyuk-kucuk donusumu
df_fake['title'] = df_fake['title'].apply(lambda x : " ".join(x.lower() for x in x.split()))
#noktamala işaretleri
df_fake['title'] = df_fake['title'].str.replace('[^\w\s]','')
#sayılar
df_fake['title'] = df_fake['title'].str.replace('\d','')
#stopwrods
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words('english')
df_fake['title'] = df_fake['title'].apply(lambda x:" ".join(x for x in x.split() if x not in sw ))
#stemming
from nltk.stem import PorterStemmer
st= PorterStemmer()
df_fake['title'] = df_fake['title'].apply(lambda x:" ".join([st.stem(i) for i in x.split()]))
#lemmi
from textblob import Word
nltk.download('wordnet')
df_fake['title'] = df_fake['title'].apply(lambda x:" ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
df_fake = df_fake[0:1298]
df_fake
```

```python
data_real =pd.read_csv("real.csv")
data_real["status"]="real"
data_real.dropna()
df_real = pd.DataFrame()
df_real["title"]= data_real["headline_text"].apply(str)
df_real["status"] = data_real["status"]
df_real
```

```python
#buyuk-kucuk donusumu
df_real['title'] = df_real['title'].apply(lambda x : " ".join(x.lower() for x in x.split()))
#noktamala işaretleri
df_real['title'] = df_real['title'].str.replace('[^\w\s]','')
#sayılar
df_real['title'] = df_real['title'].str.replace('\d','')
#stopwrods
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words('english')
df_real['title'] = df_real['title'].apply(lambda x:" ".join(x for x in x.split() if x not in sw ))
#stemming
from nltk.stem import PorterStemmer
st= PorterStemmer()
df_real['title'] = df_real['title'].apply(lambda x:" ".join([st.stem(i) for i in x.split()]))
#lemmi
from textblob import Word
nltk.download('wordnet')
df_real['title'] = df_real['title'].apply(lambda x:" ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
df_real
```

```python
df_real = df_real[0:1968]
wholedata = pd.concat([df_fake,df_real], axis = 0).reset_index()
wholedata = wholedata.drop(columns=['index'])
wholedata.dropna()
wholedata
```

```python
x_train, x_test , y_train, y_test = model_selection.train_test_split(wholedata["title"],wholedata["status"],test_size = 0.15, random_state=1)
x_train, x_val, y_train, y_val = model_selection.train_test_split(x_train,y_train,test_size = 0.175, random_state=1)
```

```python
encoder = preprocessing.LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_val = encoder.fit_transform(y_val)
y_test = encoder.fit_transform(y_test)
```

```python
tf_idf_word_vectorizer = TfidfVectorizer()
tf_idf_word_vectorizer.fit(x_train)
x_train_tf_idf_word = tf_idf_word_vectorizer.transform(x_train)
x_val_tf_idf_word = tf_idf_word_vectorizer.transform(x_val)
x_test_tf_idf_word = tf_idf_word_vectorizer.transform(x_test)
```

```python
#Lojistik Regresyon
loj = linear_model.LogisticRegression(C=30)
loj_model = loj.fit(x_train_tf_idf_word,y_train)
prediction_loj = loj_model.predict(x_val_tf_idf_word)
print(accuracy_score(y_val,prediction_loj))
print(model_selection.cross_val_score(loj_model,x_val_tf_idf_word,y_val,cv=10).mean())
print(confusion_matrix(y_val, prediction_loj))
prediction_loj = loj_model.predict(x_test_tf_idf_word)
print(accuracy_score(y_test,prediction_loj))
print(model_selection.cross_val_score(loj_model,x_test_tf_idf_word,y_test,cv=10).mean())

```
